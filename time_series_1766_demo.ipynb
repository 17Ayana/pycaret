```python
import numpy as np
import matplotlib.pyplot as plt

from sktime.datasets import load_airline
from sktime.transformations.series.scaledlogit import ScaledLogitTransformer
```

```python
y = load_airline()

# set of limits
limits_set =[
    [0.0,625],
    [0.0,600],
    [20,625],
    [None,None],
    [None,600],
    [None,500]
]
# matplotlib subplots
fig, axes = plt.subplots(nrows=len(limits_set), figsize=(10,20))
fig.patch.set_facecolor('white')
# loop over limits
for index,limits in enumerate(limits_set):
    limiter= ScaledLogitTransformer(*limits)
    limiter.fit(y)
    axes[index].plot(limiter.fit_transform(y).values)
    axes[index].set_title(f"Limits: {limits}")
fig.show()
```
